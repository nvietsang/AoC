```python
f = open("puzzle-08-input.txt", "r")
data = f.read()
f.close()
data = data.splitlines()
grid = [[int(x) for x in line] for line in data]
nrows = len(data[0])
ncols = len(data)

print(nrows)
print(ncols)
```

```python

def check(i, j) -> bool:
    is_top_visible = True
    for i1 in range(i):
        if grid[i1][j] >= grid[i][j]:
            is_top_visible = False
            break 
    if is_top_visible:
        return True

    is_bottom_visible = True
    for i1 in range(i+1, nrows):
        if grid[i1][j] >= grid[i][j]:
            is_bottom_visible = False
            break    
    if is_bottom_visible:
        return True

    is_left_visible = True
    for j1 in range(j):
        if grid[i][j1] >= grid[i][j]:
            is_left_visible = False
    if is_left_visible:
        return True

    if_right_visible = True
    for j1 in range(j+1, ncols):
        if grid[i][j1] >= grid[i][j]:
            if_right_visible = False    
    if if_right_visible:
        return True
    
    return False

n_visible_trees = 0
for i in range(1, nrows-1):
    for j in range(1, ncols-1):
        if check(i,j):
            n_visible_trees += 1

n_visible_trees += (2*nrows + 2*ncols - 4)
n_visible_trees
```

```python
def check(i, j) -> bool:
    top_score = 0
    for i1 in range(i-1, -1, -1):
        if grid[i1][j] >= grid[i][j]:
            top_score = i - i1
            break
        if i1 == 0:
            top_score = i

    # if i == 1 and j == 1:
    #     print(f"T: {top_score}") 

    bottom_score = 0
    for i1 in range(i+1, nrows):
        if grid[i1][j] >= grid[i][j]:
            bottom_score = i1 - i
            break
        if i1 == nrows-1:
            bottom_score = nrows-i-1    

    # if i == 1 and j == 1:
    #     print(f"B: {bottom_score}") 

    left_score = 0
    for j1 in range(j-1, -1, -1):
        if grid[i][j1] >= grid[i][j]:
            left_score = j - j1
            break
        if j1 == 0:
            left_score = j

    # if i == 1 and j == 1:
    #     print(f"L: {left_score}") 

    right_score = 0
    for j1 in range(j+1, ncols):
        if grid[i][j1] >= grid[i][j]:
            right_score = j1 - j
            break
        if j1 == ncols-1:
            right_score = ncols-j-1

    # if i == 1 and j == 1:
    #     print(f"R: {right_score}") 

    return top_score * bottom_score * left_score * right_score


highest_scenic_score = 0
for i in range(nrows):
    for j in range(ncols):
        new_score = check(i, j)
        if highest_scenic_score < new_score:
            highest_scenic_score = new_score

highest_scenic_score
```
