```python
f = open("puzzle-09-test.txt", "r")
data = f.read()
f.close()
data = data.splitlines()
data
```

```python
xH = 0
yH = 0
xT = 0
yT = 0

class Position:
    OL = (0, 0)
    LL = (1, 0)
    RR = (-1, 0)
    TT = (0, -1)
    BB = (0, 1)
    TL = (1, -1)
    TR = (-1, -1)
    BL = (1, 1)
    BR = (-1, 1)

position_visited = [(xT, yT)]

for line in data:
    line = line.split()
    move = line[0]
    step = int(line[1])

    for s in range(step):
        match (xH-xT, yH-yT):
            case Position.OL:
                match move:
                    case "L":
                        xH -= 1
                    case "R":
                        xH += 1
                    case "U":
                        yH += 1
                    case "D":
                        yH -= 1
                    case _:
                        raise ValueError
            case Position.LL:
                match move:
                    case "L":
                        xH -= 1
                    case "R":
                        xH += 1
                        xT += 1
                        position_visited.append((xT, yT))
                    case "U":
                        yH += 1
                    case "D":
                        yH -= 1
                    case _:
                        raise ValueError
            case Position.RR:
                match move:
                    case "L":
                        xH -= 1
                        xT -= 1
                        position_visited.append((xT, yT))
                    case "R":
                        xH += 1
                    case "U":
                        yH += 1
                    case "D":
                        yH -= 1
                    case _:
                        raise ValueError
            case Position.TT:
                match move:
                    case "L":
                        xH -= 1
                    case "R":
                        xH += 1
                    case "U":
                        yH += 1
                    case "D":
                        yH -= 1
                        yT -= 1
                        position_visited.append((xT, yT))
                    case _:
                        raise ValueError
            case Position.BB:
                match move:
                    case "L":
                        xH -= 1
                    case "R":
                        xH += 1
                    case "U":
                        yH += 1
                        yT += 1
                        position_visited.append((xT, yT))
                    case "D":
                        yH -= 1
                    case _:
                        raise ValueError
            case Position.TL:
                match move:
                    case "L":
                        xH -= 1
                    case "R":
                        xH += 1
                        xT += 1
                        yT -= 1
                        position_visited.append((xT, yT))
                    case "U":
                        yH += 1
                    case "D":
                        yH -= 1
                        xT += 1
                        yT -= 1
                        position_visited.append((xT, yT))
                    case _:
                        raise ValueError
            case Position.TR:
                match move:
                    case "L":
                        xH -= 1
                        xT -= 1
                        yT -= 1
                        position_visited.append((xT, yT))
                    case "R":
                        xH += 1
                    case "U":
                        yH += 1
                    case "D":
                        yH -= 1
                        xT -= 1
                        yT -= 1
                        position_visited.append((xT, yT))
                    case _:
                        raise ValueError
            case Position.BL:
                match move:
                    case "L":
                        xH -= 1
                    case "R":
                        xH += 1
                        xT += 1
                        yT += 1
                        position_visited.append((xT, yT))
                    case "U":
                        yH += 1
                        xT += 1
                        yT += 1
                        position_visited.append((xT, yT))                    
                    case "D":
                        yH -= 1
                    case _:
                        raise ValueError
            case Position.BR:
                match move:
                    case "L":
                        xH -= 1
                        xT -= 1
                        yT += 1
                        position_visited.append((xT, yT))
                    case "R":
                        xH += 1
                    case "U":
                        yH += 1
                        xT -= 1
                        yT += 1
                        position_visited.append((xT, yT))
                    case "D":
                        yH -= 1
                    case _:
                        raise ValueError
            case _:
                raise ValueError

position_visited
```

```python
f = open("puzzle-09-input.txt", "r")
data = f.read()
f.close()
data = data.splitlines()
data

x = [0] * 10
y = [0] * 10

pos9_visited = [(x[9], y[9])]

def move_head(xH, yH, move):
    match move:
        case "L":
            xH -= 1
        case "R":
            xH += 1
        case "U":
            yH += 1
        case "D":
            yH -= 1
        case _  :
            raise ValueError
    return (xH, yH)

def move_tail(xH, yH, xT, yT):
    match (xH-xT, yH-yT):
        case (1, 2):
            xT += 1
            yT += 1
        case (0, 2):
            yT += 1
        case (-1, 2):
            xT -= 1
            yT += 1

        case (-2, 1):
            xT -= 1
            yT += 1
        case (-2, 0):
            xT -= 1
        case (-2, -1):
            xT -= 1
            yT -= 1

        case (-1, -2):
            xT -= 1
            yT -= 1
        case (0, -2):
            yT -= 1
        case (1, -2):
            xT += 1
            yT -= 1

        case (2, -1):
            xT += 1
            yT -= 1
        case (2, 0):
            xT += 1
        case (2, 1):
            xT += 1
            yT += 1

        case (-2, 2):
            xT -= 1
            yT += 1
        case (-2, -2):
            xT -= 1
            yT -= 1
        case (2, -2):
            xT += 1
            yT -= 1
        case (2, 2):
            xT += 1
            yT += 1
        case _ :
            pass
    return (xT, yT)


for line in data:
    line = line.split(" ")
    move = line[0]
    step = int(line[1])

    for s in range(step):
        # move H
        x[0], y[0] = move_head(x[0], y[0], move)
        # move 9 tails
        for i in range(1, 10):
            x[i], y[i] = move_tail(x[i-1], y[i-1], x[i], y[i])
            if i == 9:
                pos9_visited.append((x[i], y[i]))   

                
print(x)
print(y)
```
